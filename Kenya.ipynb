# 4. Highly Correlated Columns Extraction
Created `correlation_df` to store all column correlations.
Generated a DataFrame listing each column and others correlated at` ≥ 0.85`.
Cleaned this DataFrame by removing redundant entries like `"Year"` and `"parsed year"`.
Saved the cleaned correlations as `highly_correlated_df_cleaned.xlsx.`

```python
# prompt: create a whole new dataframe that shows how all these columns are correlated

# Calculate the correlation matrix
correlation_matrix = kenya.corr()

# Create a new DataFrame from the correlation matrix
correlation_df = pd.DataFrame(correlation_matrix)

# Display the new DataFrame
correlation_df
```

```python
# prompt: using this "correlation_df" i want a df that shows a column as a row and the columns that it is highly correlated to as the cells, the number of cells may vary but I want all columns that have a 0.85+ correlation to the column the row is representing. kindly note that I want names of the related columns on the cell not the correlation value. i expect a df that has a column name in its first column and the columns it is highly correlated to following it in the next columns

import pandas as pd

# Assuming 'correlation_df' is already defined from the previous code

def get_highly_correlated_columns(correlation_df, threshold=0.85):
    highly_correlated_df = pd.DataFrame()
    for col in correlation_df.columns:
        highly_correlated_cols = correlation_df[col][abs(correlation_df[col]) >= threshold].index.tolist()
        # Remove the column itself from the list of highly correlated columns
        highly_correlated_cols = [c for c in highly_correlated_cols if c != col]

        if highly_correlated_cols:  # Check if there are any highly correlated columns
          row_data = [col] + highly_correlated_cols
          row_df = pd.DataFrame([row_data])
          highly_correlated_df = pd.concat([highly_correlated_df, row_df], ignore_index = True)

    return highly_correlated_df

highly_correlated_df = get_highly_correlated_columns(correlation_df)

#Display the resulting DataFrame
highly_correlated_df
```

# 6. Lagged Correlations
Focused on economic metrics `(GDP and Military Expenditure)`.
Created lagged GDP growth over 5 years.
Examined correlations between:
Lagged GDP growth `(GDP_Growth)` and current `"Military expenditure (% of GDP)"`.
Visualized the relationship with scatter plots.

```python
import pandas as pd

# Load the cleaned dataset
kenya = pd.read_excel('kenya_cleaned.xlsx')
```

```python
# prompt: output the columns of the kenya df

kenya.columns
```

```python
import matplotlib.pyplot as plt# Convert relevant columns to numeric, handling errors
kenya['GDP (current US$)'] = pd.to_numeric(kenya['GDP (current US$)'], errors='coerce')
kenya['Military expenditure (% of GDP)'] = pd.to_numeric(kenya['Military expenditure (% of GDP)'], errors='coerce')

# Create lagged GDP growth
kenya['GDP_Growth'] = kenya['GDP (current US$)'].pct_change() * 100  # Calculate percentage change

# Lag the GDP growth by 5 years
kenya['Lagged_GDP_Growth'] = kenya['GDP_Growth'].shift(5)

# Drop rows with NaN values created by the lag
kenya = kenya.dropna(subset=['Lagged_GDP_Growth', 'Military expenditure (% of GDP)'])


# Calculate the correlation between lagged GDP growth and current military expenditure
correlation = kenya['Lagged_GDP_Growth'].corr(kenya['Military expenditure (% of GDP)'])

print(f"Correlation between lagged GDP growth (previous 5 years) and current military expenditure: {correlation}")

# Visualize the relationship (optional)
plt.figure(figsize=(10, 6))
plt.scatter(kenya['Lagged_GDP_Growth'], kenya['Military expenditure (% of GDP)'])
plt.xlabel('Lagged GDP Growth (previous 5 years)')
plt.ylabel('Military Expenditure (% of GDP)')
plt.title('Correlation between Lagged GDP Growth and Military Expenditure')
plt.grid(True)
plt.show()
```

# **Final insights generated from visualizations**

# 1. Gross National Income (GNI) Trends
**2008 Decline:** The notable drop in GNI in 2008 correlates with the global financial crisis originating in the United States. The recession in the US led to reduced global demand for exports, increased credit constraints, and a general economic slowdown. Kenya, as part of the global economy, faced reduced trade volumes, lower remittances, and a tightening of foreign investments, significantly affecting its economic output.

**2007 Growth Rate Reduction:** The slowdown in GNI growth in 2007 can be primarily attributed to Kenya's post-election violence. The unrest disrupted businesses, transportation, and agricultural activities—sectors that significantly contribute to the economy. This led to reduced productivity, lower investor confidence, and a direct hit to GNI even before the global crisis.
# 2. Life Expectancy Growth

**Steady Increase (2000–2015):** Life expectancy rose from 54 years in 2000 to nearly 62 years in 2015. This improvement is strongly linked to the expansion of healthcare facilities and infrastructure under President Mwai Kibaki's leadership.

**Healthcare Advancements:** Increased access to medical services, better immunization coverage, and reduced infant and maternal mortality rates significantly contributed to this upward trend.

**Improved Living Standards: **Economic growth likely facilitated better nutrition through more balanced diets and improved hygiene practices, enhancing the population's overall health and longevity.
# 3. Fertility Rate Decline
Drop from 5.1 to 3.8 (2000–2015):

**Cost of Living:** The rising cost of living made larger families financially unsustainable. Unlike pre-independence times, children now require education and healthcare, increasing parental responsibilities and expenses.

**Increased Contraceptive Use:** Greater awareness and access to family planning services led to a widespread adoption of contraceptives, directly influencing fertility rates.

**Modernization:** The gradual modernization of Kenyan society played a significant role. Urbanized and educated women are more likely to prioritize smaller family sizes, career opportunities, and personal development over traditional norms.

**Cultural Shifts:** Changes in societal expectations about family size and child-rearing responsibilities influenced the trend.
