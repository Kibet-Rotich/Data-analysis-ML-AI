**Data cleaning**

I will proceed to fill null values with the mean and also drop duplicates.

```python
df.isnull().sum()
```

As you can see alot of the age is null, but since its a integer, I'll fill it with the mean. Considering the cabin I will use the front fill function to do the same.

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Cabin'].fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

I will proceed to drop the null values for now.

```python
df.dropna(inplace=True)
df.isnull().sum()
```

**Data Analysis**

This is what we have been waiting for. let us answer some questions.

1. What factors had the highest influence on passenger survival?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze the relationship between survival and other features
# e.g., Pclass, Sex, Age, Fare, Embarked

# 1. Survival rate by Passenger Class
print(df[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean())
sns.barplot(x='Pclass', y='Survived', data=df)
plt.show()

# 2. Survival rate by Sex
print(df[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean())
sns.barplot(x='Sex', y='Survived', data=df)
plt.show()

# 3. Survival rate by Age
# You could create age groups and analyze survival within those groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 18, 40, 60, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])
print(df[['AgeGroup', 'Survived']].groupby(['AgeGroup'], as_index=False).mean())
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.show()

# 4. Survival rate by Fare
# You could create fare groups and analyze survival within those groups
df['FareGroup'] = pd.qcut(df['Fare'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
print(df[['FareGroup', 'Survived']].groupby(['FareGroup'], as_index=False).mean())
sns.barplot(x='FareGroup', y='Survived', data=df)
plt.show()

# 5. Survival rate by Embarked
print(df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean())
sns.barplot(x='Embarked', y='Survived', data=df)
plt.show()

# Based on the analysis, you can identify the factors with the highest influence on survival.
# For example, passenger class, sex, and age appear to be significant factors.
```

From the looks of things, I think gender played a big role, with 74% probability of females surviving. You can also use a corelation matrix, but this only works for integers.

```python
correlation_matrix = df[['Survived', 'Pclass', 'Age', 'Fare']].corr()
print(correlation_matrix)
```

2. How did survival rates vary across different ticket classes?

3. Was there any relationship between fare and survival?

```python
df[['Fare', 'Survived']].groupby(['Survived']).mean().plot(kind='bar')
plt.show()
```

We can see here that on average, higher paying passengers had a better chance of survival.

4. Did the port of embarkation affect survival rates?

```python
df[['Embarked', 'Survived']].groupby(['Embarked']).mean().plot(kind='bar')
plt.show()
```

People who boarded on port C had better chances of survival

5. How did family size affect survival?

```python
familysize = df['SibSp'] + df['Parch'] + 1
df['FamilySize'] = familysize
df[['FamilySize', 'Survived']].groupby(['FamilySize']).mean().plot(kind='bar')
```

Larger family sizes typically with 5 members and above had dramatically decreased chances of survival. Being in a family with 3 or 4 members increased your chances or survival.
