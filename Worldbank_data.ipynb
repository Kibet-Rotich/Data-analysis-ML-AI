# **Correlation Heatmap for Kenya's Indicators**

The goal of this section was to identify relationships between various indicators:

Since the original data had years as columns and indicators as rows, I transposed the data to make indicators columns for correlation analysis.
Using Seaborn, I plotted a correlation heatmap to visually identify relationships between indicators like GDP, population, and military expenditure.
The heatmap highlights potential associations, such as high correlation values between certain indicators, guiding further analysis.

```python
kenya['Year'] = kenya['Year'].str.extract(r'(\d{4})', expand=False)
kenya['Year'] = pd.to_numeric(kenya['Year'], errors='coerce').astype('Int64') # Convert to numeric, handle errors, then to nullable integer
kenya['Year'] = pd.to_datetime(kenya['Year'], format='%Y')

kenya
```

```python
# prompt: I want to plot a correlation heatmap for the kenya df, i want to know which columns are related, only problem is that the columns are the years and what I want to know if they are related are in the rows, can you change that?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data (assuming 'kenya_data.xlsx' exists from previous code)
kenya = pd.read_excel('kenya_data.xlsx')

# Data preprocessing (similar to what you've done, but more robust)
kenya = kenya[kenya['Country Name'] == 'Kenya']
kenya = kenya.drop(['Country Name', 'Country Code', 'Series Code'], axis=1)
kenya = kenya.set_index('Series Code').transpose()

# Convert year columns to numeric, handling non-numeric values
for col in kenya.columns:
    kenya[col] = pd.to_numeric(kenya[col], errors='coerce')

# Calculate the correlation matrix
correlation_matrix = kenya.corr()


# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Indicators for Kenya')
plt.show()
```
